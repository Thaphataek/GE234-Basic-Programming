## 🔹 ตัวอย่างที่ 1: การใช้ `for loop` เพื่อแสดงค่าพิกัด


```python

# พิกัดของสถานที่สำคัญในประเทศไทย
locations = {
    "Bangkok": (13.7563, 100.5018),
    "Chiang Mai": (18.7883, 98.9853),
    "Phuket": (7.8804, 98.3923),
    "Ubon Ratchathani": (15.2287, 104.8564)
}

# วนลูปแสดงค่าพิกัด
for city, coord in locations.items():
    print(f"{city}: Latitude {coord[0]}, Longitude {coord[1]}")
```


## 🔹 ตัวอย่างที่ 2: การใช้ `if-else` กรองข้อมูลพิกัด


```python

for city, coord in locations.items():
    if coord[0] > 16:
        print(f"{city} อยู่ทางภาคเหนือของประเทศไทย")
    else:
        print(f"{city} ไม่ใช่จังหวัดทางภาคเหนือ")
```


## 🔹 ตัวอย่างที่ 3: ใช้ `while loop` นับจำนวนจุดข้อมูล


```python

points = [(13.7563, 100.5018), (18.7883, 98.9853), (7.8804, 98.3923)]
count = 0

while count < len(points):
    print(f"Point {count + 1}: {points[count]}")
    count += 1
```


## 🔹 ตัวอย่างที่ 4: วิเคราะห์ข้อมูล Raster ด้วย `for loop`


```python

import numpy as np

# จำลองข้อมูล NDVI (Normalized Difference Vegetation Index)
ndvi = np.array([
    [0.2, 0.4, 0.5],
    [0.1, 0.3, 0.7],
    [0.6, 0.8, 0.2]
])

# วนลูปตรวจสอบค่าพืชพรรณ
for row in range(ndvi.shape[0]):
    for col in range(ndvi.shape[1]):
        if ndvi[row, col] > 0.3:
            print(f"พิกัด ({row}, {col}) มีพืชพรรณหนาแน่น (NDVI={ndvi[row, col]})")
```

### คำอธิบาย: เรื่องสี

*   เราสร้างคลาส `Colors` เพื่อเก็บรหัส ANSI escape codes สำหรับแต่ละสี เพื่อให้เรียกใช้งานได้ง่ายขึ้นและโค้ดดูสะอาดตา
*   `'\033'` คือ escape character ที่บอกว่ากำลังจะเริ่มต้นรหัสควบคุม
*   `'[91m'` ถึง `'[97m'` คือรหัสสำหรับสีข้อความต่างๆ
*   `'[101m'` และ `[102m` คือรหัสสำหรับสีพื้นหลัง (background color)
*   `'\033[0m'` คือรหัส `RESET` ที่ใช้ในการรีเซ็ตสีกลับไปเป็นค่าเริ่มต้น เพื่อไม่ให้สีที่กำหนดไว้ส่งผลต่อข้อความถัดไป

เพียงแค่นำรหัสสีไปไว้หน้าข้อความที่คุณต้องการเปลี่ยนสี และตามด้วย `Colors.RESET` หลังข้อความนั้นๆ ค่ะ

```python
# กำหนดรหัสสี ANSI
class Colors:
    RESET = '\033[0m'
    RED = '\033[91m'
    GREEN = '\033[92m' # แก้ไขกลับเป็นสีเขียวสด
    YELLOW = '\033[93m'
    BLUE = '\033[94m'
    CYAN = '\033[96m'
    WHITE = '\033[97m'
    MAGENTA = '\033[95m'

    # สีแบ็คกราวด์
    BG_RED = '\033[101m'
    BG_GREEN = '\033[102m'

# ตัวอย่างการใช้งาน
print(f"ปกติ: นี่คือข้อความสี {Colors.RED}แดง{Colors.RESET}")
print(f"ปกติ: นี่คือข้อความสี {Colors.GREEN}เขียว{Colors.RESET}")
print(f"ปกติ: นี่คือข้อความสี {Colors.YELLOW}เหลือง{Colors.RESET}")
print(f"ปกติ: นี่คือข้อความสี {Colors.BLUE}น้ำเงิน{Colors.RESET}")
print(f"ปกติ: นี่คือข้อความสี {Colors.CYAN}ฟ้าอมเขียว{Colors.RESET}")
print(f"ปกติ: นี่คือข้อความสี {Colors.MAGENTA}ม่วงแดง{Colors.RESET}")
print(f"ปกติ: นี่คือข้อความสี {Colors.WHITE}ขาว{Colors.RESET}")
print(f"ปกติ: นี่คือข้อความสี {Colors.BG_RED}พื้นหลังแดง{Colors.RESET}")
# บรรทัดที่เคยมีปัญหา Colors.BLACK ถูกลบออกเนื่องจากตอนนี้ไม่มีแล้ว
```

```python
#ข้อ2
locations = {
    "Phu Kradueng": (16.8713889, 101.7186111, 1316), # coord[2] ความสูง หน่วยเมตร
    "Wat Phra Kaew": (13.7503447, 100.4912523 , 2),
    'Doi Inthanon': ( 18.5886801, 98.4873894, 2565),
    'Prang Sam Yot': (14.8029016, 100.6140149 , 17),
    "Thammasat University Lampang Campus": (18.3174441, 99.3992693 , 254),
    'Prince of Songkla University' : (6.8776002, 101.2367824 , 6),
    'Khao Kitchakut' : (12.8367121 , 102.1686578 , 1003)

}

for city, coord in locations.items():
    if coord[2]> 600:    #โดยทั่วไปภูเขาจะถูกกำหนดให้มีความสูงตั้งแต่ 600 เมตร (ประมาณ 2,000 ฟุต) ขึ้นไปจากพื้นที่
        print(f"{city}.......{Colors.RED}พื้นที่ภูเขา{Colors.RESET}")
    else:
        print(f"{city} .......{Colors.GREEN}พื้นที่ราบ{Colors.RESET}")
```

```python
#ข้อ3
points = [(16.8713889), (13.7503447), (18.5886801), (14.8029016), (18.3174441), (6.8776002), (12.8367121)]
count = 0

print("แสดงค่าละติจูดจนกว่าจะพบค่าต่ำกว่า 10°:")
while count < len(points):
    current_latitude = points[count]

    print(f"\nPoint {count + 1}: Latitude {current_latitude}°")
    if current_latitude < 10:

        print(f"พบค่าละติจูดต่ำกว่า 10° ที่ {current_latitude}° หยุดการวนลูป")
        break # หยุดการทำงานของ while loop
    count += 1
```
