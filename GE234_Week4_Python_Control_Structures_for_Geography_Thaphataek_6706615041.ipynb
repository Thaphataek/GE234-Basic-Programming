**เกณฑ์:**
- NDVI < 0.2 → พื้นที่เปลือย/น้ำ
- 0.2 ≤ NDVI < 0.4 → พืชพรรณน้อย
- 0.4 ≤ NDVI < 0.6 → พืชพรรณปานกลาง
- NDVI ≥ 0.6 → ป่าไม้/พืชพรรณหนาแน่น

**ให้เขียนโปรแกรม** รับค่า NDVI และแสดงประเภทพืชพรรณ

# ส่วนที่ 2: คำสั่งวนซ้ำ for loop

for loop ใช้สำหรับวนซ้ำผ่านข้อมูลใน list, tuple, string หรือ range

## 2.1 วนซ้ำใน list

````python
# ตัวอย่าง: แสดงอุณหภูมิทั้ง 8 วัน
print("อุณหภูมิรายวัน:")
for temp in temperatures:
    print(f"  🌡️ {temp}°C")
````

## 2.2 คำนวณค่าเฉลี่ย

````python
# ตัวอย่าง: คำนวณค่าเฉลี่ยอุณหภูมิ
total = 0 #เริ่มตั้งแต่ 0
count = 0 #จำนวน n ที่เอามาหาร

for temp in temperatures:
    total += temp
    count += 1

average = total / count
print(f"อุณหภูมิเฉลี่ย: {average:.2f}°C")
````

## 2.3 วนซ้ำด้วย range()

````python
# ตัวอย่าง: แสดงวันที่ 1-10
print("วันที่ในเดือน:")
for day in range(1, 11):  # 1-10 เวลาเขียนต้องเพิ่มวันไป +1
    print(f"  วันที่ {day}")
````

## 2.4 วนซ้ำพร้อมเงื่อนไข

````python
# ตัวอย่าง: นับจำนวนวันที่ฝนตก
rainy_days = 0

for rain in rainfall:
    if rain == 0:
        rainy_days += 1

print(f"จำนวนวันที่ไม่ฝนตก: {rainy_days} วัน")
````

````python
# ตัวอย่าง: หาวันที่อุณหภูมิสูงที่สุด
max_temp = temperatures[0]
max_day = 1

for i in range(len(temperatures)): #len คือ length แปลว่าระยะทาง
    if temperatures[i] > max_temp:
        max_temp = temperatures[i]
        max_day = i + 1

print(f"วันที่ {max_day} มีอุณหภูมิสูงที่สุด {max_temp}°C")
````

## 2.5 ใช้ enumerate() - แสดงทั้งลำดับและค่า

````python
# ตัวอย่าง: แสดงวันและอุณหภูมิพร้อมกัน
print("อุณหภูมิรายวัน:")
for  day, temp in enumerate(temperatures, start=1):
    print(f"  วันที่ {day}: {temp}°C")
````

# ส่วนที่ 3: คำสั่งวนซ้ำ while loop

while loop จะวนซ้ำไปเรื่อยๆ ตราบใดที่เงื่อนไขยังเป็นจริง

## 3.1 ตัวอย่างพื้นฐาน

````python
# ตัวอย่าง: นับจำนวนวันจนครบ 10 วัน
day = 1

while day <= 10:
    print(f"วันที่ {day}")
    day += 1  # เพิ่มค่า day

print("ครบ 10 วันแล้ว!")
````

## 3.2 while loop กับเงื่อนไข

````python
# ตัวอย่าง: จำลองการอ่านอุณหภูมิจนกว่าจะเกิน 35 องศา
temps_sample = [28, 30, 32, 34, 36, 38]
i = 0

while i < len(temps_sample) and temps_sample[i] <= 35:
    print(f"วันที่ {i+1}: อุณหภูมิ {temps_sample[i]}°C - ปกติ")
    i += 1

if i < len(temps_sample):
    print(f"⚠️ วันที่ {i+1}: อุณหภูมิ {temps_sample[i]}°C - สูงเกินเกณฑ์!")
````

## 3.3 ความแตกต่าง for vs while

**ใช้ for loop เมื่อ:**
- รู้จำนวนรอบที่ต้องวนแน่นอน
- ต้องการวนซ้ำผ่านข้อมูลใน list, dictionary

**ใช้ while loop เมื่อ:**
- ไม่รู้จำนวนรอบที่แน่นอน
- ต้องการวนซ้ำจนกว่าเงื่อนไขจะเป็นเท็จ
- สร้าง menu หรือ interactive program

## 📝 แบบฝึกหัดที่ 6: การเพิ่มพื้นที่ป่าไม้

**โจทย์:** เริ่มต้นมีพื้นที่ป่า 5,000 ไร่ และเพิ่มขึ้นปีละ 500 ไร่

**ให้เขียนโปรแกรม** ใช้ while loop นับว่าต้องใช้เวลากี่ปีจึงจะมีพื้นที่ป่า 10,000 ไร่

````python
# แบบฝึกหัดที่ 6: เขียนโค้ดของคุณที่นี่
forest_area = 5000  # ไร่
target = 10000      # ไร่
years = 0

# เขียนโค้ดตรงนี้
while forest_area < target:
    forest_area += 500  # เพิ่มพื้นที่ป่า 500 ไร่ต่อปี
    years += 1          # นับจำนวนปี

print(f"ต้องใช้เวลา {years} ปี ในการเพิ่มพื้นที่ป่าให้ถึง {target} ไร่")
````

# ส่วนที่ 4: การควบคุม Loop (break และ continue)

- **break**: หยุด loop ทันที
- **continue**: ข้ามรอบนี้ไปรอบถัดไป

## 4.1 break - หยุด loop ทันที

````python
# ตัวอย่าง: หยุดทันทีที่พบ PM2.5 เกิน 150 (อันตราย)
print("ตรวจสอบคุณภาพอากาศ:")
for day, pm in enumerate(pm25, start=1):
    if pm >= 150:
        print(f"🔴 วันที่ {day}: PM2.5 = {pm} - อันตราย! หยุดการตรวจสอบ")
        break
    else:
        print(f"วันที่ {day}: PM2.5 = {pm}")
````

## 4.2 continue - ข้ามรอบนี้

````python
# ตัวอย่าง: แสดงเฉพาะวันที่มีฝนตก (ข้ามวันที่ไม่มีฝน)
rainfall = [110, 0, 32, 21, 0 ,27, 0, 0, 132, 15]
print("วันที่ฝนตก:")
for day, rain in enumerate(rainfall, start=1):
    if rain == 0:
        continue  # ข้ามวันที่ไม่มีฝน
    print(f"  🌧️ วันที่ {day}: ฝน {rain} mm")
````

# ส่วนที่ 5: Nested Loops (Loop ซ้อน Loop)

Loop ซ้อนกันใช้สำหรับข้อมูลที่มีหลายมิติ เช่น ตาราง, grid

## 5.1 ตัวอย่างง่ายๆ

````python
# ตัวอย่าง: แสดงตารางอุณหภูมิ 3 เมือง x 5 วัน
cities = ['กรุงเทพ', 'เชียงใหม่', 'ภูเก็ต']
temps_by_city = [
    [32, 33, 31, 34, 32],  # กรุงเทพ
    [28, 29, 27, 30, 28],  # เชียงใหม่
    [30, 31, 30, 32, 31]   # ภูเก็ต
]

print("อุณหภูมิรายเมือง:")
for i in range(len(cities)):
    print(f"\n{cities[i]}:")
    for day in range(len(temps_by_city[i])):
        print(f"  วันที่ {day+1}: {temps_by_city[i][day]}°C")
````

## 5.2 ประยุกต์ใช้กับ Grid

````python
# ตัวอย่าง: วิเคราะห์ข้อมูล grid 3x3 (เหมือน raster)
grid = [
    [45, 52, 48],
    [55, 120, 60],  # ค่ากลางสูงมาก!
    [50, 58, 53]
]

print("PM2.5 Grid (3x3):")
max_value = 0
max_row = 0
max_col = 0

for row in range(3):
    for col in range(3):
        value = grid[row][col]
        print(f"  [{row},{col}] = {value}", end="  ")

        # หาค่าสูงสุด
        if value > max_value:
            max_value = value
            max_row = row
            max_col = col
    print()  # ขึ้นบรรทัดใหม่

print(f"\nค่าสูงสุด: {max_value} ที่ตำแหน่ง [{max_row},{max_col}]")
````

## 📝 แบบฝึกหัดที่ 9: วิเคราะห์ PM2.5 หลายเมือง

**โจทย์:** มีข้อมูล PM2.5 ของ 3 เมือง 7 วัน

**ให้เขียนโปรแกรม:**
1. แสดงข้อมูล PM2.5 ทุกเมืองทุกวัน
2. คำนวณค่าเฉลี่ย PM2.5 ของแต่ละเมือง
3. หาเมืองที่มีค่าเฉลี่ยสูงที่สุด

```python
cities = ['กรุงเทพ', 'เชียงใหม่', 'ขอนแก่น']
pm_by_city = [
    [45, 50, 55, 48, 52, 60, 58],  # กรุงเทพ
    [85, 90, 120, 95, 100, 88, 92], # เชียงใหม่

]
```

````python
# แบบฝึกหัดที่ 9: เขียนโค้ดของคุณที่นี่
cities = ['กรุงเทพ', 'เชียงใหม่', 'ขอนแก่น']
pm_by_city = [
    [45, 50, 55, 48, 52, 60, 58],
    [85, 90, 120, 95, 100, 88, 92],
    [55, 60, 65, 58, 62, 70, 68]
]

# เขียนโค้ดตรงนี้
max_avg_pm = -1
worst_city = ""

print("=== วิเคราะห์ PM2.5 รายเมือง ===")
for i in range(len(cities)):
    city = cities[i]
    pm_values = pm_by_city[i]

    print(f"\nเมือง: {city}")
    total_pm = 0

    for day, pm in enumerate(pm_values, start=1):
        print(f"  วันที่ {day}: PM2.5 = {pm}")
        total_pm += pm

    average_pm = total_pm / len(pm_values)
    print(f"  ค่าเฉลี่ย PM2.5: {average_pm:.2f}")

    if average_pm > max_avg_pm:
        max_avg_pm = average_pm
        worst_city = city

print(f"\n=== สรุป ===")
print(f"เมืองที่มีค่าเฉลี่ย PM2.5 สูงที่สุดคือ: {worst_city} (ค่าเฉลี่ย {max_avg_pm:.2f})")
````

# ส่วนที่ 6: 🎯 Mini Project - ระบบวิเคราะห์คุณภาพอากาศ

## โจทย์โปรเจกต์

สร้างโปรแกรมวิเคราะห์คุณภาพอากาศที่สามารถ:

### ข้อมูลที่ให้:
- 5 จังหวัด: กรุงเทพ, เชียงใหม่, ขอนแก่น, ภูเก็ต, อุดรธานี
- ข้อมูล PM2.5 รายวัน 7 วัน

### โปรแกรมต้องทำได้:
1. **แสดงข้อมูล PM2.5** ทุกจังหวัดทุกวัน พร้อมจำแนกระดับคุณภาพอากาศ
2. **คำนวณค่าเฉลี่ย PM2.5** ของแต่ละจังหวัด
3. **หาจังหวัดที่มีคุณภาพอากาศแย่ที่สุด** (ค่าเฉลี่ยสูงสุด)
4. **นับจำนวนวันที่อากาศดี** (PM2.5 < 50) ทั้งหมด
5. **สรุปรายงาน** ในรูปแบบที่อ่านง่าย

### เกณฑ์จำแนกคุณภาพอากาศ:
- PM2.5 ≤ 50: ดี 🟢
- 51-100: ปานกลาง 🟡
- 101-150: แย่ 🟠
- > 150: อันตราย 🔴

````python
# ข้อมูลสำหรับโปรเจกต์
provinces = ['กรุงเทพ', 'เชียงใหม่', 'ขอนแก่น', 'ภูเก็ต', 'อุดรธานี']

pm25_data = [
    [45, 52, 48, 55, 50, 58, 53],    # กรุงเทพ
    [95, 110, 125, 105, 115, 98, 102], # เชียงใหม่
    [62, 68, 70, 65, 72, 69, 71],    # ขอนแก่น
    [38, 42, 40, 45, 43, 48, 41],    # ภูเก็ต
    [78, 85, 82, 88, 80, 92, 86]     # อุดรธานี
]

print("✅ ข้อมูลพร้อมแล้ว! เริ่มเขียนโค้ดได้เลย")
````

## 💡 คำแนะนำ (Hints)

1. ใช้ **nested loop** วนผ่านทุกจังหวัดและทุกวัน
2. ใช้ **if-elif-else** จำแนกระดับคุณภาพอากาศ
3. สร้าง **list** เก็บค่าเฉลี่ยของแต่ละจังหวัด
4. ใช้ตัวแปรนับจำนวนวันที่อากาศดี
5. ค่อยๆ ทำทีละส่วน ทดสอบไปเรื่อยๆ

## ตัวอย่าง Output ที่ต้องการ:

```
=== วิเคราะห์คุณภาพอากาศ 5 จังหวัด ===

กรุงเทพ:
  วันที่ 1: PM2.5 = 45 🟢 ดี
  วันที่ 2: PM2.5 = 52 🟡 ปานกลาง
  ...
  ค่าเฉลี่ย: 51.57

เชียงใหม่:
  วันที่ 1: PM2.5 = 95 🟡 ปานกลาง
  ...

สรุปผล:
- จังหวัดที่มีคุณภาพอากาศแย่ที่สุด: เชียงใหม่ (ค่าเฉลี่ย 107.14)
- จำนวนวันที่อากาศดี: 8 วัน
```

````python
# 🎯 Mini Project: เขียนโค้ดของคุณที่นี่

print("=" * 50)
print("=== วิเคราะห์คุณภาพอากาศ 5 จังหวัด ===")
print("=" * 50)

# ตัวแปรสำหรับเก็บผลลัพธ์รวม
max_avg_pm = -1
worst_province = ""
good_air_days_count = 0

# วนลูปผ่านแต่ละจังหวัด
for i in range(len(provinces)):
    province = provinces[i]
    pm_values = pm25_data[i]

    print(f"\n{province}:")
    total_pm_province = 0

    # วนลูปผ่านข้อมูล PM2.5 ในแต่ละวันของจังหวัดนั้นๆ
    for day, pm in enumerate(pm_values, start=1):
        # จำแนกระดับคุณภาพอากาศและนับวันที่อากาศดี
        if pm <= 50:
            status = "🟢 ดี"
            good_air_days_count += 1
        elif pm <= 100:
            status = "🟡 ปานกลาง"
        elif pm <= 150:
            status = "🟠 แย่"
        else:
            status = "🔴 อันตราย"
        print(f"  วันที่ {day}: PM2.5 = {pm} {status}")
        total_pm_province += pm

    # คำนวณค่าเฉลี่ย PM2.5 ของจังหวัดนั้นๆ
    average_pm_province = total_pm_province / len(pm_values)
    print(f"  ค่าเฉลี่ย: {average_pm_province:.2f}")

    # ตรวจสอบว่าจังหวัดนี้มีค่าเฉลี่ยแย่ที่สุดหรือไม่
    if average_pm_province > max_avg_pm:
        max_avg_pm = average_pm_province
        worst_province = province

print("\n" + "=" * 50)
print("=== สรุปผล ===")
print("=" * 50)
print(f"- จังหวัดที่มีคุณภาพอากาศแย่ที่สุด: {worst_province} (ค่าเฉลี่ย {max_avg_pm:.2f})")
print(f"- จำนวนวันที่อากาศดี (PM2.5 < 50) ทั้งหมด: {good_air_days_count} วัน")


print("\n" + "=" * 50)
print("จบการวิเคราะห์")
print("=" * 50)
````
